#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hours = input('Enter Hours: ')
rate = input('Enter Rate: ')
if int(hours)>40 :
  regularpay = 40 * int (rate)
  overtimerate= (int (hours)-40 ) * (int (rate)*1.5)
  overtimepay = overtimerate + regularpay
  print("Pay: ",overtimepay)
else:
  pay= int(hours)*int(rate)
  print ("Pay: ",pay)
````

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

Adapted from [PEW] E3.3
<br>

````python
score = int (input ('Enter Score: '))
if score>=97 and score <= 100 :
  print('Grade: 1.0')
elif score>=93 and score <=97 :
  print('Grade: 1.25')
elif score>=89 and score <=93 :
  print('Grade: 1.5')
elif score>=85 and score <=89 :
  print('Grade: 1.75')
elif score>=81 and score <=85 :
  print('Grade: 2.0')
elif score>=77 and score <=81 :
  print('Grade: 2.25')
elif score>=74 and score <=77 :
  print('Grade: 2.5')
else:
  print('Grade: 3.0')
````
